```python
import numpy as np

# 1) Hours,Minutes,Seconds to Radians

def hms_to_rads(hr,m,s):
    '''
    converts given hours,minutes,seconds to decimal radians
    args:
    hr: hour component
    m: minutes component
    s: seconds component
    
    returns rads, value in radians
    '''
    deg = (hr + m/60 + s/3600) * 15
    rads = deg*(np.pi/180)
    
    #print(str(hr)+'h'+str(m)+'m'+str(s)+'s is '+str(np.round(rads,4))+' radians')
    
    return rads

rads1 = hms_to_rads(3,20,45)

###############################################
# 2) Degrees,Arcminutes,Arcseconds to Radians

def dms_to_rads(deg,m,s):
    '''
    converts given degrees,arcminutes,arcseconds to decimal radians
    args:
    deg: degrees component
    m: arcminutes component
    s: arcseconds component
    
    returns rads, value in radians
    '''
    just_deg = deg+m/60+s/3600
    rads = just_deg*(np.pi/180)
    
    #print(str(deg)+"\u00b0"+str(m)+"'"+str(s)+'"'+' is '+str(np.round(rads,4))+' radians')
    
    return rads

rads2 = dms_to_rads(30,40,45)

######################################
# 3) Radians to Hours,Minutes,Seconds

def rads_to_hms(rads):
    '''
    converts given radians to hours,minutes,seconds
    args:
    rads: decimal radians
    
    
    returns hr,m,s
    '''
    deg = rads * (180/np.pi) * 3600
    
    all_seconds = deg*(1/15)
    
    hr = (all_seconds // 3600)
    m = ((all_seconds % 3600) // 60)
    s = (all_seconds % 60)
    
    #print(str(np.round(rads,4))+' radians is '+str(int(hr))+'h'+str(int(m))+'m'+str(int(s))+'s')
    
    return hr,m,s

hms3 = rads_to_hms(rads1) #gives roughly same answer as input for #1

#############################################
# 4) Radians to Degrees,Arcminutes,Arcseconds

def rads_to_dms(rads):
    '''
    converts given radians to degrees,arcminutes,arcseconds
    args:
    rads: decimal radians
    
    
    returns deg,m,s
    '''
    all_arcseconds = rads * (180/np.pi) * 3600
    
    
    deg = (all_arcseconds // 3600)
    m = ((all_arcseconds % 3600) // 60)
    s = (all_arcseconds % 60)
    
    #print(str(np.round(rads,4))+' radians is '+str(int(deg))+"\u00b0"+str(int(m))+"'"+str(int(s))+'"')
    
    return deg,m,s



dms4 = rads_to_dms(rads2) # gives answer same as input for #2





```
